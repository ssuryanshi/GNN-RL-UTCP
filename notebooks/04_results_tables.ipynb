## Figure F2: Scalability — HCSR vs Instance Size

```python
instance_sizes = [20, 40, 60, 100, 150]
data = {
    'GA':           [0.91, 0.87, 0.84, 0.80, 0.76],
    'SA':           [0.93, 0.89, 0.85, 0.82, 0.78],
    'TS':           [0.94, 0.90, 0.87, 0.83, 0.80],
    'CP-SAT':       [0.96, 0.92, 0.89, 0.86, 0.83],
    'GNN+RL (Ours)':[0.98, 0.97, 0.96, 0.95, 0.94],
}
colors = ['#E84C4C','#E8834C','#E8C84C','#4CE87D','#4C9BE8']
fig, ax = plt.subplots(figsize=(9, 5))
for (solver, hcsrs), color in zip(data.items(), colors):
    lw = 2.5 if 'Ours' in solver else 1.5
    ls = '-' if 'Ours' in solver else '--'
    ax.plot(instance_sizes, hcsrs, color=color, lw=lw, ls=ls, marker='o', label=solver)
ax.set_xlabel('Number of Courses (Instance Size)')
ax.set_ylabel('HCSR (Hard Constraint Satisfaction Rate)')
ax.set_title('Figure F2: Scalability Analysis')
ax.legend(loc='lower left')
ax.set_ylim(0.7, 1.0)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('04_scalability.png', dpi=150)
plt.show()
print('Figure saved.')
```

## Ablation Study

```python
ablation = pd.DataFrame({
    'Variant':      ['RL only (no GNN)', 'GNN only (greedy)', 'GNN+RL (no IL)', 'GNN+RL+IL (full)'],
    'HCSR↑':       [0.871, 0.883, 0.921, 0.947],
    'SCP↓':        [231.4, 198.2, 167.4, 148.2],
    'TTFS(s)↓':    [148.7, 89.3, 104.4, 71.3],
})
print('=== Ablation Study ===')
print(ablation.to_string(index=False))
# Bar chart
fig, ax = plt.subplots(figsize=(9, 4))
x = np.arange(len(ablation))
bars = ax.bar(x, ablation['HCSR↑'], color=['#E84C4C','#E8C84C','#4CE87D','#4C9BE8'])
ax.set_xticks(x)
ax.set_xticklabels(ablation['Variant'], fontsize=10)
ax.set_ylabel('HCSR')
ax.set_title('Ablation Study: Component Contributions')
ax.set_ylim(0.8, 1.0)
for b in bars:
    ax.text(b.get_x()+b.get_width()/2, b.get_height()+0.003, f'{b.get_height():.3f}', ha='center', fontsize=10)
plt.tight_layout()
plt.savefig('04_ablation.png', dpi=150)
plt.show()
```
