# Notebook 03: Training Curves
Plot reward vs timesteps, comparing cold-start RL vs IL warm-start RL.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import uniform_filter1d
plt.rcParams.update({'font.size': 13, 'axes.spines.top': False, 'axes.spines.right': False})
```

## Simulated Training Curves
After actual training runs, replace `rewards_cold` and `rewards_il` with logged data from `checkpoints/monitor.csv`.

```python
np.random.seed(42)
n_steps = 500
timesteps = np.linspace(0, 5_000_000, n_steps)

# Simulated cold-start RL
rewards_cold = -300 + 350 * (1 - np.exp(-timesteps / 2_500_000))
rewards_cold += np.random.normal(0, 30, n_steps)
rewards_cold_smooth = uniform_filter1d(rewards_cold, size=25)

# Simulated IL warm-start
rewards_il = -100 + 370 * (1 - np.exp(-timesteps / 1_200_000))
rewards_il += np.random.normal(0, 20, n_steps)
rewards_il_smooth = uniform_filter1d(rewards_il, size=25)
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(timesteps/1e6, rewards_cold, alpha=0.2, color='#E84C4C')
ax.plot(timesteps/1e6, rewards_cold_smooth, color='#E84C4C', linewidth=2, label='PPO (cold start)')
ax.plot(timesteps/1e6, rewards_il, alpha=0.2, color='#4C9BE8')
ax.plot(timesteps/1e6, rewards_il_smooth, color='#4C9BE8', linewidth=2, label='GNN+PPO (IL warm-start)')
ax.axhline(0, color='k', linewidth=0.8, linestyle='--')
ax.set_xlabel('Timesteps (million)')
ax.set_ylabel('Episode Reward')
ax.set_title('Training Curves: Cold Start vs IL Warm-Start (F1)')
ax.legend()
ax.set_xlim(0, 5)
plt.tight_layout()
plt.savefig('03_training_curves.png', dpi=150)
plt.show()
print('Figure saved.')
```

## HCSR During Training

```python
hcsr_cold = 0.6 + 0.3 * (1 - np.exp(-timesteps / 3_000_000)) + np.random.normal(0, 0.02, n_steps)
hcsr_il   = 0.75 + 0.22 * (1 - np.exp(-timesteps / 1_500_000)) + np.random.normal(0, 0.015, n_steps)
hcsr_cold = np.clip(uniform_filter1d(hcsr_cold, 25), 0, 1)
hcsr_il   = np.clip(uniform_filter1d(hcsr_il, 25), 0, 1)
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(timesteps/1e6, hcsr_cold, color='#E84C4C', linewidth=2, label='PPO (cold start)')
ax.plot(timesteps/1e6, hcsr_il,   color='#4C9BE8', linewidth=2, label='GNN+PPO (IL warm-start)')
ax.set_xlabel('Timesteps (million)')
ax.set_ylabel('HCSR')
ax.set_ylim(0.55, 1.0)
ax.set_title('Hard Constraint Satisfaction Rate During Training')
ax.legend()
plt.tight_layout()
plt.savefig('03_hcsr_training.png', dpi=150)
plt.show()
```
